```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

#initialize tokenizer and model from pretrained GPT2 model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
#sequence : your words, doesn't need to follow this
#sequence="He began his premiership by forming a five-man war cabiner which included Chamerlain as Lord President"
sequence="Neelu is wearing glasses and she is 5 feet and 5 inch tall"
inputs = tokenizer.encode(sequence, return_tensors='pt')
outputs = model.generate(inputs, max_length=200, do_sample=True, temperature=1)

tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
import torch
from transformers import BertTokenizer, BertLMHeadModel


#initialize tokenizer and model from pretrained GPT2 model
berttokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bertmodel = BertLMHeadModel.from_pretrained('bert-base-uncased')
```

```python
#sequence="He began his premiership by forming a five-man war cabiner which included Chamerlain as Lord President"
sequence="Artificial intelligence is intelligence—perceiving, synthesizing, and inferring information—demonstrated by machines, as opposed to intelligence displayed by humans or by other animals"
inputs = berttokenizer.encode(sequence, return_tensors='pt')
outputs = bertmodel.generate(inputs, max_length=200, do_sample=True, temperature=1)

berttokenizer.decode(outputs[0], skip_special_tokens=True)
```
